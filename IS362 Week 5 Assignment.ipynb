#### 1.What is the northernmost airport in the United States?
#### 2.What is the easternmost airport in the United States?
#### 3.On February 12th, 2013, which New York area airport had the windiest weather?

```python
# Pulling Airports data from Github link into a DF
airportsdata = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/airports.csv')
```

```python
# Specifying what columns are needed in airports DF
airportsdf = pd.DataFrame(
    airportsdata,
    columns=[
        'faa',
        'name',
        'lat',
        'lon',
        'alt',
        'tz',
        'dst',
        'tzone'
    ],
)
```

```python
print(airportsdf)
```

```python
# Sorting DF by 'lat' (Latitude) by descending order to have northnermost locations, selecting top 5 entries.
airportsdf.sort_values(by='lat', ascending=False).head(5)
```

## 1. The northernmost airport is Dillant Hopkins Airport.

```python
# Sorting DF by 'lon' (Longitude) by descending order to have easternmost locations, selecting top 5 entries.
airportsdf.sort_values(by='lon', ascending=False).head(5)
```

```python
# Pulling Weather data from Github link into a DF
weatherdata = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/weather.csv')
```

```python
# Specifying what columns are needed in weather DF
weatherdf = pd.DataFrame(
    weatherdata,
    columns=[
        'origin',
        'year',
        'month',
        'day',
        'hour',
        'temp',
        'dewp',
        'humid',
        'wind_dir',
        'wind_speed',
        'wind_gust',
        'precip',
        'pressure',
        'visib',
        'time_hour'
    ],
)
```

```python
print(weatherdf)
```

```python
#2/12/2013
requesteddatedf = weatherdf[
    (weatherdf['year'].isin(['2013'])) & 
    (weatherdf['month'].isin(['2'])) & 
     (weatherdf['day'].isin(['13']))
     ]
```

```python
print(requesteddatedf)
```

```python
# Sorting DF by 'wind_speed' by descending order to have the windiest locations, selecting top 5 entries.
requesteddatedf.sort_values(by='wind_speed', ascending=False).head(5)
```
